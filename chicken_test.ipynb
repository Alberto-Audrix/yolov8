```python
import os

# Tentukan direktori image_list
image_list_dir = "dataset_sementara/dead_chicken"

# Inisialisasi list untuk menyimpan path gambar
image_paths = []

# Iterasi melalui semua file dalam direktori
for filename in os.listdir(image_list_dir):
    # Periksa apakah file adalah file gambar
    if filename.endswith(('.jpg', '.jpeg', '.png', '.gif')):
        # Dapatkan path lengkap dari file gambar
        image_path = os.path.join(image_list_dir, filename)
        # Tambahkan path gambar ke list
        image_paths.append(image_path)

# Print semua path gambar
for path in image_paths:
    print(path)
```

```python
from ultralytics import YOLO
from PIL import Image


# Create a new YOLO model from scratch
model = YOLO('yolov8n.yaml')

# Load a pretrained YOLO model (recommended for training)
model = YOLO('runs/detect/train(80,20,10)_70_epoch/weights/best.pt')

# im1 = Image.open("images/sample4.jpg")
# im1 = Image.open("image_list")

for paths in image_paths:
    im1 = Image.open(paths)
    results = model.predict(source=im1, save=True, save_txt = True)
    



# results = model.predict(source=im1, save=True, save_txt = True)
# print(results)

# for box in result.boxes:
#   class_id = result.names[box.cls[0].item()]
#   cords = box.xyxy[0].tolist()
#   cords = [round(x) for x in cords]
#   conf = round(box.conf[0].item(), 2)
#   print("Object type:", class_id)
#   print("Coordinates:", cords)
#   print("Probability:", conf)
#   print("---")
```
